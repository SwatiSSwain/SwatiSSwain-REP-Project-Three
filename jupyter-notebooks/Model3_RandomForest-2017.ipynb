```python
indicators_df = pd.read_sql("SELECT * from table5_2017", connection)
                              
indicators_df.head() 
```

```python
pivot_df = indicators_df.pivot(index='countryname', columns='indicatorname', values='year_2017')
```

```python
pivot_df.head()
```

```python
pivot_df=pivot_df.dropna(how='any')
```

```python
pivot_df.count()
```

```python
target_groups_df = pd.read_sql("SELECT distinct countryname,target_groups  from table5_2017", connection)
                              
target_groups_df.head() 
```

```python
merged_df = pd.merge(pivot_df, target_groups_df 
                   ,left_on="countryname"
                   ,right_on = "countryname"
                   ,how="inner")
```

```python
merged_df.head()
```

```python
target = merged_df["target_groups"]
target.head()
```

```python
target_names = pd.read_sql("SELECT distinct target_groups  from table5_2017", connection)
target_names
```

```python
data = merged_df.drop(["target_groups","countryname"], axis=1)
```

```python
feature_names = data.columns
data.head()
```

```python
data.count()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
